```julia
x = range(0, 10, length = 100)
y1 = sin.(x)
y2 = cos.(x)
plot(x,[y1,y2])
```

```julia
x = range(0, 10, length = 100)
y1 = sin.(x)
y2 = cos.(y1)
plot(x,[y1,y2], title = "trigonometric functions", label = ["sin(x)" "cos(y1)"])
y3 = @. cos.(x) 
plot!(x,y3)
```

```julia
savefig("plot-1.png")
```

```julia
x = range(0, 10, length = 100)
y = sin.(x)
noise = @. sin(x) + 0.1*randn()
plot(x,y, label = "sin(x)", lw = 3)
plot!(x,noise, label = "noise")
```

```julia
x = range(0,10,length=100)
y = sin.(x)
noise = @. sin(x) + 0.1*randn()
plot(x,y,label = "sinx")
scatter!(x,noise,label = "data", ma = 0.9)
title!("sin with noise")
xlabel!("x")
ylabel!("y")
plot!(legend=:bottomleft)
```

```julia
savefig("plot-2.png")
```

```julia
using Plots
x = randn(10^5)
yrange = range(-5,5,length = 21)
histogram(x, bins = yrange, color =:grey)
```

```julia
savefig("histogram.png")
```

```julia
p(x) =  1/sqrt(2pi) * exp(-x^2/2)
y = range(-5,5, length = 21)
histogram(x, label = "experimental", normalize=:true, bins = y, color=:grey)
plot!(p, label = "analytical", lw = 2)
xlims!(-5,5)
ylims!(0,0.5)
xlabel!("x")
ylabel!("p(x)")
```

```julia
savefig("normalization.png")
```
